```python
# importar las librerias
import pandas as pd
import numpy as np
```

```python
#importo el dataset
JuegosOlimpicos = pd.read_csv("C:/Users/Romi Ott/CoderHouse Ejercicios/Summer-Olympic-medals-1976-to-2008.csv", encoding = "Latin-1")
JuegosOlimpicos
```

1) construir un objeto de series de tiempo con el índice igual al año de los juegos olímpicos

```python
#primero, genero el indice por año
fechasOlimp = JuegosOlimpicos["Year"]
indice=fechasOlimp

#luego, elimino la columna de año del dataframe 
JuegosSinAnio = JuegosOlimpicos.drop(["Year"], axis=1, inplace=False)
columnas = JuegosSinAnio.columns[:]

#finalmente, genero el nuevo dataframe con el indice por año
JuegosPorAnio = pd.DataFrame(JuegosSinAnio.values,index=indice,columns=columnas)
JuegosPorAnio
```

2) Separar nombres de apellidos en dos columnas distintas usando la coma como separador

```python
#primero, separo los nombres y apellidos
JuegosOlimpicos[["AthleteName","AthleteSurName"]] = JuegosOlimpicos["Athlete"].str.split(",", n=1, expand=True)

#luego, los integro a los datos
JuegosOlimpicos.drop(["Athlete"], axis=1, inplace=True)
JuegosOlimpicos
```

3) Obtener medidas resumen del conjunto de datos. Cuál es el país que ganó más medallas?

```python
#genero la base de datos con agrupacion de la informacion de mi interes
MF = JuegosOlimpicos.groupby(["Country", "Medal"]).count()
MF

#ordeno los datos de mayor a menor y selecciono un criterio de visualizacion
MFOrdenados = MF.sort_values(by=["Sport"], ascending=False)
MFOrdenados[["Sport"]].head(10)

#RTA: Estados Unidos gano mas medallas en todas las categorias
    #en segundo lugar, la (ex)Union Sovietica con mas medallas de oro ganadas
    #seguidos de Australia, con la mayor cantidad de medallas de bronce
```
