2) Realizar un histograma con los salarios. Qué rango(s) de salarios son los más populares?

```python
#veo la mediana, para saber si la tendencia de los datos es centrada o no (hay un desvio de los datos hacia la izquierda)
RecursosHumanos.Salary.median()
```

```python
#luego, grafico la misma para que sea mas sencillo de interpretar
HistoSalarios = sns.histplot(data=RecursosHumanos, x="Salary", stat="proportion")
plt.axvline(x=RecursosHumanos.Salary.mean(), color="red")
plt.axvline(x=RecursosHumanos.Salary.median(), color="green")

#RTA: Dada la distribucion de los datos, entre el 25% y el 75% de los salarios se encuentran entre $55.501 y $72.036
#De estos, los salarios mas populares son $57.815, $61.242 y $63.025, encontrandse dentro del rango descripto anteriormente.
```

3) Realizar un gráfico de violín con los salarios, discriminados por género.

```python
#para el grafico de violin, se debe tener en cuenta que el genero esta en formato 0 y 1, donde el 0=Femenino y 1=Masculino.
ViolinSalarios = sns.catplot(data=RecursosHumanos, kind='violin', x="GenderID", y="Salary", split=False)
ViolinSalarios.set_xticklabels(["0=Female","1=Male"])

#Rta: Si bien los sueldos no se distribuyen de igual manera, puede verse que la mediana de ambos es la misma.
#Esto se da en casos donde la distribucion esta balanceada en cuanto al recuento (o sea, el 50% de los sueldos esta sobre esa mediana
#y el otro 50% del otro lado). Esto no implica que sean similares, como se ve en las distribuciones.
#A nivel salarial, las mujeres presentan una diversidad salarial mayor, ya que existen unas pocas empleadas con sueldos muy altos
#y el fuerte de la fuerza laboral esta cercano a la mediana. En el caso de los hombres, los salarios son mas cercanos a 
#la mediana, sin presentar sueldos maximos tan altos como los de las mujeres, pero si una cantidad de hombres con sueldos
#altos mas significativa, notandose en el ancho de la curva en la parte superior.
```
